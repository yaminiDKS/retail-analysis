# What are the top-selling items during each season?Is there a pattern in the purchase behavior based on the season?

```python
new_df = df[['Item Purchased','Category','Purchase Amount (USD)','Season']].copy()
new_df.head()
```

```python
res = new_df.groupby('Season').size()
res
```

All season have the same number of purchases

```python
res1 = new_df.groupby(['Season','Item Purchased']).size()
res1
```

```python
mean_counts = res1.groupby(level='Season').mean()
```

```python
order = ['Spring', 'Summer', 'Fall', 'Winter']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for ax, season in zip(axes.flatten(), order):
    season_data = res1.loc[season]
    season_mean = mean_counts.loc[season]
    std_dev = season_data.std()

    bars = season_data.plot(kind='bar', ax=ax, color='skyblue')

    upper_bound = season_mean + std_dev * 1.28
    lower_bound = season_mean - std_dev * 1.28
    for bar, count in zip(bars.patches, season_data):
        if count > upper_bound:
            bar.set_color('green')
        if count < lower_bound:
            bar.set_color('red')

    ax.axhline(y=season_mean, color='red', linestyle='--', label='Mean')
    ax.axhline(y=upper_bound, color='blue', linestyle='--', label='Upper Bound (90% CI)')
    ax.axhline(y=lower_bound, color='blue', linestyle='--', label='Lower Bound (90% CI)')
    ax.set_title(f'Season: {season} ({season_mean:.2f})')
    ax.set_xlabel('Item Purchased')
    ax.set_ylabel('Count')

fig.suptitle('Item Purchased Count for Each Season', fontsize=20)
plt.tight_layout()
plt.show()
```

**Spring:**
- **Sales Variations:**
  - **Decreased Sales:**
    - Hats, Jeans, Pants
  - **Increased Sales:**
    - Sweater, Shorts

- **Reasoning:**
  - **Decreased Sales:**
    - Hats, Jeans, and Pants may experience reduced demand as customers transition to lighter and more casual clothing options during the spring.
  - **Increased Sales:**
    - Sweaters and Shorts see increased sales, likely due to the changing weather and the desire for versatile pieces suitable for both cool and warm temperatures.
- **Pattern:**
   - During spring, customers show an inclination towards purchasing lighter and versatile clothing items like Sweaters and Shorts, while demand decreases for heavier items such as Hats, Jeans, Pants, and Sunglasses.

**Summer:**
- **Sales Variations:**
  - **Increased Sales:**
    - Pants, Jewellery, Dress
  - **Decreased Sales:**
    - Skirt, Sweater, Gloves, Hoodies, Jackets, Jeans

- **Reasoning:**
  - **Increased Sales:**
    - Pants, Jewellery, and Dresses witness higher demand as customers embrace lighter and more fashionable options suitable for summer.
  - **Decreased Sales:**
    - Skirts, Sweaters, Jackets,Hoodies,Jeaans and Gloves may experience decreased sales as customers opt for cooler alternatives during the warmer summer months.

- **Pattern:**
  - In summer, there is a surge in the demand for light and fashionable items like Pants, Jewellery, Dresses, and Shoes. Conversely, sales decrease for warmer clothing and accessories such as Skirts, Sweaters, Gloves, Hoodies, Jackets, and Jeans.

**Fall:**
- **Sales Variations:**
  - **Decreased Sales:**
    - Shoes, Sneakers
  - **Increased Sales:**
    - Handbag, Jacket, Hat

- **Reasoning:**
  - **Decreased Sales:**
    - Shoes and Sneakers see reduced demand as customers transition to closed footwear options like Boots during the fall.
  - **Increased Sales:**
    - Handbags, Jackets, and Hats witness increased sales as customers prepare for cooler temperatures and seek stylish fall accessories.

 **Pattern:**
   - Fall sees increased demand for stylish accessories like Handbags, Jackets, and Hats, while sales decrease for certain footwear items like Shoes and Sneakers.

**Winter:**
- **Sales Variations:**
  - **Increased Sales:**
    - Sunglasses, Shirts, Pants, Hoodie
  - **Decreased Sales:**
    - Jeans, Backpack

- **Reasoning:**
  - **Increased Sales:**
    - Sunglasses, Shirts, Pants, and Hoodies experience higher demand during winter, reflecting the need for both stylish and warm clothing as customers navigate colder temperatures.
  - **Decreased Sales:**
    - Jeans and Backpacks may see reduced sales as customers prioritize other winter-specific items, possibly warmer pants and more substantial bags.

   **Pattern:**
   - Winter witnesses a rise in the demand for winter-appropriate clothing and accessories such as Shirts, Pants, and Hoodies. Sales decrease for items like Jeans and Backpacks, possibly due to a preference for heavier pants and more substantial bags during the colder season.

**General Insights:**


- **Trends:**
  - Lighter and more breathable clothing tends to sell well in warmer seasons, while heavier and warmer items are favored in colder seasons.

- **Seasonal Adaptations:**
  - Customer preferences adapt to seasonal changes, influencing the demand for specific clothing and accessories.

- **Weather-Driven Choices:**
